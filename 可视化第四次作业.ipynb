```python
import matplotlib.pyplot as plt
import numpy as np
# 中文设置
plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei"]
plt.rcParams['axes.unicode_minus'] = False
labels = ["性别", "教育程度", "对醒狮的了解程度", "对醒狮的态度", "年龄"]
sizes = [7.6, 14.3, 17.1, 27.0, 34.0]
colors = ["#86bcb0", "#b7449d", "#ff9810", "#ff6968", "#496bca"]
explode = [0.06]*5
plt.figure(figsize=(5, 5), dpi=95)
wedges, _, autotexts = plt.pie(
    sizes,
    explode=explode,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90,
    textprops={"fontsize":11, "color":"white"},
    labels=None,
    radius=0.75
)
# 添加图例
plt.title("醒狮相关调研影响因素占比", fontsize=13, pad=12)
plt.legend(wedges, labels, loc="lower right", fontsize=9)
plt.axis("equal")
plt.tight_layout()
plt.show()
```
