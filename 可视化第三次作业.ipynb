#  Q2

```python
import pandas as pd
data_dict = {
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
}
df1 = pd.DataFrame(data_dict)
print("原始 DataFrame：")
print(df1)

df_part1 = df1[["score", "gender"]]
print("\n1. 选取 score 和 gender 两列：")
print(df_part1)

df_part2 = df1.loc[[0, 2]]
print("\n2. 选取索引 0、2 的行：")
print(df_part2)

df_part3 = df1[df1["score"] > 80]
print("\n3. 筛选 score>80 的行：")
print(df_part3)

df1["grade"] = df1["score"].apply(lambda x: "优秀" if x >= 90 else "良好")
print("\n4. 新增 grade 列后：")
print(df1)

df1.drop(columns=["gender"], inplace=True)
print("\n5. 删除 gender 列后：")
print(df1)
```

# Q3

```python
import pandas as pd
data = {"x": [1, None, 3], "y": [None, 5, 6]}
df = pd.DataFrame(data)

print("1. 原始含缺失值的 DataFrame：")
print(df)

print("\n2. 缺失值检测结果（True 表示该位置为缺失值）：")
missing_check = df.isnull()
print(missing_check)

print("\n3. 每列缺失值数量统计：")
missing_count = df.isnull().sum()
print(missing_count)

df_filled = df.fillna(0)
print("\n4. 用 0 填充缺失值后的 DataFrame：")
print(df_filled)

df_dropped = df.dropna()
print("\n5. 删除含缺失值的行后的 DataFrame：")
print(df_dropped)
```

# Q4

```python
import pandas as pd
data_dict = {
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
}
df1 = pd.DataFrame(data_dict)

df_grouped = df1.groupby("gender")["score"].agg(["mean", "max"])
print("1. 按 gender 分组，score 的均值和最大值：")
print(df_grouped)

df_count = df1.groupby("gender").size()
print("\n2. 按 gender 分组，每组人数：")
print(df_count)
```

# Q5

```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
data_dict = {
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
}
df1 = pd.DataFrame(data_dict)

print("1. 原始 DataFrame：")
print(df1.to_string())

df_sorted = df1.sort_values(by="score", ascending=True)
print("\n2. 按 score 升序排序后的 DataFrame：")
print(df_sorted.to_string())

df1.loc[3] = ["Tom", 85, "男"]  
print("\n3. 添加重复行后的 DataFrame：")
print(df1.to_string())

df_deduplicated = df1.drop_duplicates()
print("\n4. 去重后的 DataFrame：")
print(df_deduplicated.to_string())
```

# Q6

```python
import pandas as pd
df1 = pd.DataFrame({"id":[1,2], "name":["X","Y"]})
df2 = pd.DataFrame({"id":[1,2], "age":[20,30]})

df_merge = pd.merge(df1, df2, on="id")
print("横向合并结果：")
print(df_merge)

df2 = pd.concat([df1, df2], axis=1)
print(df2)

df_a = pd.DataFrame({"val":[1,2]})
df_b = pd.DataFrame({"val":[3,4]})
df_concat = pd.concat([df_a, df_b], ignore_index=True)
print("\n纵向拼接并重置索引结果：")
print(df_concat)
```

# Q7

```python
import pandas as pd
df = pd.DataFrame({"date": ["2025-01-01", "2025-01-02", "2025-01-03"]})
df["date"] = pd.to_datetime(df["date"])

print("转换时间格式后：")
print(df)

df["month"] = df["date"].dt.month       
df["weekday"] = df["date"].dt.weekday    
df["weekday_name"] = df["date"].dt.day_name(locale="zh_CN")  
print("\n提取月份、星期后：")
print(df)
```

# Q8

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})
mean_score = df["score"].mean()
std_score = df["score"].std()

df["score_std"] = (df["score"] - mean_score) / std_score
print("标准化后数据：")
print(df)

pivot_df = pd.pivot_table(df, index="gender", values="score", aggfunc="median")
print("\n透视表（中位数）：")
print(pivot_df)
```
