```python
import time
import re
import csv
import pymysql
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from collections import Counter, defaultdict
# 1. 设置中文黑体
plt.rcParams["font.family"] = ["SimHei"]  # Windows 黑体
# Mac 换成：plt.rcParams["font.family"] = ["Arial Unicode MS"]
# Linux 换成：plt.rcParams["font.family"] = ["WenQuanYi Micro Hei"]

# 2. 解决负号'-'显示为方框问题
plt.rcParams["axes.unicode_minus"] = False


# =========================
# 1. 基本配置
# =========================

CITY = "xm"      # 练习 1：爬取城市，厦门是 xiamen
LIMIT = 500          # 爬取 500 条

DB_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": "123456",
    "database": "ershoufang",
    "port": 3306,
    "charset": "utf8mb4"
}

TABLE_NAME = "ershoufang_list"

# 每一个练习对应一个表格文件
HOUSE_LIST_CSV = f"{CITY}_house_list.csv"
DISTRICT_AVG_PRICE_CSV = f"{CITY}_district_avg_price.csv"
HOUSE_TYPE_COUNT_CSV = f"{CITY}_house_type_count.csv"
DISTRICT_COUNT_CSV = f"{CITY}_district_count.csv"
DISTRICT_COUNT_IMAGE = f"{CITY}_district_count.png"


FIELDS = [
    "城市", "市区", "标题", "户型", "面积", "方位", "楼层", "时间",
    "所属小区", "所属区域", "总价", "均价", "房龄"
]


# =========================
# 2. 数据库操作
# =========================

def get_conn():
    """连接 MySQL 数据库"""
    return pymysql.connect(**DB_CONFIG)


def create_table():
    """创建数据表"""
    conn = get_conn()
    cursor = conn.cursor()

    sql = f"""
    CREATE TABLE IF NOT EXISTS {TABLE_NAME} (
        id INT AUTO_INCREMENT PRIMARY KEY,
        城市 VARCHAR(64),
        市区 VARCHAR(64),
        标题 VARCHAR(255),
        户型 VARCHAR(64),
        面积 VARCHAR(64),
        方位 VARCHAR(64),
        楼层 VARCHAR(64),
        时间 VARCHAR(64),
        所属小区 VARCHAR(255),
        所属区域 VARCHAR(255),
        总价 VARCHAR(64),
        均价 VARCHAR(64),
        房龄 VARCHAR(64)
    ) CHARACTER SET utf8mb4;
    """

    cursor.execute(sql)
    conn.commit()
    cursor.close()
    conn.close()


def clear_city_data(city):
    """清空当前城市旧数据，避免福州和厦门数据混在一起"""
    conn = get_conn()
    cursor = conn.cursor()

    sql = f"DELETE FROM {TABLE_NAME} WHERE 城市 = %s"
    cursor.execute(sql, (city,))

    conn.commit()
    cursor.close()
    conn.close()

    print(f"已清空 {city} 的旧数据")


def insert_data(data):
    """插入一条房源数据"""
    conn = get_conn()
    cursor = conn.cursor()

    keys = ",".join(FIELDS)
    values = ",".join(["%s"] * len(FIELDS))

    sql = f"""
    INSERT INTO {TABLE_NAME} ({keys})
    VALUES ({values})
    """

    vals = [data.get(field, "") for field in FIELDS]

    cursor.execute(sql, vals)
    conn.commit()

    cursor.close()
    conn.close()


# =========================
# 3. 数据提取函数
# =========================

def extract_district(address_text):
    """从所属区域中提取市区"""
    text = re.sub(r"\s+", " ", address_text).strip()

    if not text:
        return ""

    parts = [
        part.strip()
        for part in re.split(r"[-－–—·/|\s]+", text)
        if part.strip()
    ]

    return parts[0] if parts else ""


def get_text_safe(element, selector):
    """安全提取文本"""
    result = element.select(selector)

    if result:
        return result[0].get_text(strip=True)

    return ""


def parse_house_item(item, city):
    """解析单条房源信息"""
    data = {}

    data["城市"] = city

    data["标题"] = get_text_safe(
        item,
        ".property-content-title-name"
    )

    data["所属小区"] = get_text_safe(
        item,
        ".property-content-info-comm-name"
    )

    data["所属区域"] = get_text_safe(
        item,
        ".property-content-info-comm-address"
    )

    data["市区"] = extract_district(data["所属区域"])

    data["总价"] = get_text_safe(
        item,
        ".property-price-total"
    )

    data["均价"] = get_text_safe(
        item,
        ".property-price-average"
    )

    info_items = item.select(".property-content-info-text")
    info_texts = [i.get_text(strip=True) for i in info_items]

    data["户型"] = info_texts[0] if len(info_texts) > 0 else ""
    data["面积"] = info_texts[1] if len(info_texts) > 1 else ""
    data["方位"] = info_texts[2] if len(info_texts) > 2 else ""
    data["楼层"] = info_texts[3] if len(info_texts) > 3 else ""
    data["时间"] = info_texts[4] if len(info_texts) > 4 else ""

    data["房龄"] = ""

    return data


# =========================
# 4. 爬虫主函数
# =========================

def start_driver():
    """启动 Chrome 浏览器"""
    chrome_options = Options()

    chrome_options.add_experimental_option(
        "excludeSwitches",
        ["enable-automation"]
    )

    # 建议不要 detach=True，避免看到旧的福州页面
    # chrome_options.add_experimental_option("detach", True)

    chrome_options.add_argument("--no-sandbox")

    # 无痕模式，减少浏览器缓存和城市 Cookie 影响
    chrome_options.add_argument("--incognito")

    # 禁用定位，避免根据定位跳转城市
    chrome_options.add_argument("--disable-geolocation")

    driver = webdriver.Chrome(options=chrome_options)

    return driver


def crawl_city(city, limit):
    """练习 1：爬取指定城市数据"""
    print(f"\n正在爬取城市：{city}")

    create_table()
    clear_city_data(city)

    driver = start_driver()

    inserted_count = 0

    try:
        for y in range(1, 5):

            for p in range(1, 100):

                if inserted_count >= limit:
                    print(f"已达到限制：{inserted_count}")
                    return

                url = f"https://{city}.anjuke.com/sale/p{p}-y{y}/?from=fangjia"

                print("准备打开：", url)

                driver.get(url)

                time.sleep(3)

                print("实际打开：", driver.current_url)

                soup = BeautifulSoup(driver.page_source, "html.parser")

                house_list = soup.select(".property")

                if not house_list:
                    print("当前页面没有找到房源，可能网页结构变化或出现验证码。")
                    continue

                for item in house_list:

                    if inserted_count >= limit:
                        print(f"已达到限制：{inserted_count}")
                        return

                    data = parse_house_item(item, city)

                    if not data["标题"]:
                        continue

                    insert_data(data)

                    inserted_count += 1

                    print(f"已插入第 {inserted_count} 条：{data['标题']}")

                time.sleep(1)

    finally:
        driver.quit()

    print(f"本次新增：{inserted_count}")


# =========================
# 5. 导出原始房源数据表
# =========================

def export_house_list_csv(city):
    """导出当前城市原始房源数据表"""
    conn = get_conn()
    cursor = conn.cursor()

    cursor.execute(
        f"SELECT * FROM {TABLE_NAME} WHERE 城市 = %s",
        (city,)
    )

    rows = cursor.fetchall()
    columns = [desc[0] for desc in cursor.description]

    with open(HOUSE_LIST_CSV, "w", newline="", encoding="utf-8-sig") as f:
        writer = csv.writer(f)
        writer.writerow(columns)
        writer.writerows(rows)

    cursor.close()
    conn.close()

    print(f"\n练习 1 表格已生成：{HOUSE_LIST_CSV}")
    print(f"共导出 {len(rows)} 条 {city} 房源数据")


def read_house_list_csv():
    """读取原始房源数据表"""
    with open(HOUSE_LIST_CSV, "r", encoding="utf-8-sig") as f:
        reader = csv.DictReader(f)
        return list(reader)


# =========================
# 6. 练习 2：统计不同市区平均总价
# =========================

def parse_total_price(price_text):
    """
    把总价文本转换成数字，单位统一为“万”

    示例：
    "300万" -> 300
    "300 万" -> 300
    "1.2亿" -> 12000
    """
    if not price_text:
        return None

    text = str(price_text).strip()

    if text in ["N/A", "暂无", "面议", "-", ""]:
        return None

    text = text.replace(" ", "").replace(",", "")

    match = re.search(r"(\d+(\.\d+)?)", text)

    if not match:
        return None

    price = float(match.group(1))

    if "亿" in text:
        price = price * 10000

    return price


def export_district_avg_price_table(city):
    """练习 2：生成不同市区平均总价表"""
    rows = read_house_list_csv()

    district_prices = defaultdict(list)

    for row in rows:
        if row.get("城市") != city:
            continue

        district = row.get("市区", "").strip()

        if not district:
            district = "未识别"

        total_price_text = row.get("总价", "")
        total_price = parse_total_price(total_price_text)

        if total_price is None:
            continue

        district_prices[district].append(total_price)

    result = []

    for district, prices in district_prices.items():
        count = len(prices)
        avg_price = sum(prices) / count
        max_price = max(prices)
        min_price = min(prices)

        result.append({
            "市区": district,
            "平均总价_万": round(avg_price, 2),
            "房源数": count,
            "最高总价_万": round(max_price, 2),
            "最低总价_万": round(min_price, 2)
        })

    result.sort(key=lambda x: x["平均总价_万"], reverse=True)

    with open(DISTRICT_AVG_PRICE_CSV, "w", newline="", encoding="utf-8-sig") as f:
        fieldnames = [
            "市区",
            "平均总价_万",
            "房源数",
            "最高总价_万",
            "最低总价_万"
        ]

        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(result)

    print(f"\n练习 2 表格已生成：{DISTRICT_AVG_PRICE_CSV}")

    print("\n练习 2：不同市区平均总价")
    for item in result:
        print(
            item["市区"],
            item["平均总价_万"],
            "万，房源数：",
            item["房源数"]
        )


# =========================
# 7. 练习 3：统计不同户型数量
# =========================

def export_house_type_count_table(city):
    """练习 3：生成不同户型数量表"""
    rows = read_house_list_csv()

    counter = Counter()

    for row in rows:
        if row.get("城市") != city:
            continue

        house_type = row.get("户型", "").strip()

        if not house_type:
            house_type = "未识别"

        counter[house_type] += 1

    result = []

    for house_type, count in counter.most_common():
        result.append({
            "户型": house_type,
            "数量": count
        })

    with open(HOUSE_TYPE_COUNT_CSV, "w", newline="", encoding="utf-8-sig") as f:
        fieldnames = ["户型", "数量"]

        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(result)

    print(f"\n练习 3 表格已生成：{HOUSE_TYPE_COUNT_CSV}")

    print("\n练习 3：不同户型数量")
    for item in result:
        print(item["户型"], item["数量"])


# =========================
# 8. 练习 4：市区房源数量表 + 柱状图
# =========================

def export_district_count_table_and_chart(city):
    """练习 4：生成市区房源数量表，并绘制柱状图"""
    rows = read_house_list_csv()

    counter = Counter()

    for row in rows:
        if row.get("城市") != city:
            continue

        district = row.get("市区", "").strip()

        if not district:
            district = "未识别"

        counter[district] += 1

    result = []

    for district, count in counter.most_common():
        result.append({
            "市区": district,
            "房源数量": count
        })

    with open(DISTRICT_COUNT_CSV, "w", newline="", encoding="utf-8-sig") as f:
        fieldnames = ["市区", "房源数量"]

        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(result)

    print(f"\n练习 4 表格已生成：{DISTRICT_COUNT_CSV}")

    if not result:
        print("没有可绘制的数据")
        return

    districts = [item["市区"] for item in result]
    counts = [item["房源数量"] for item in result]

    plt.figure(figsize=(10, 6))
    plt.bar(districts, counts)

    plt.title(f"{city} 各市区房源数量")
    plt.xlabel("市区")
    plt.ylabel("房源数量")

    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.savefig(DISTRICT_COUNT_IMAGE, dpi=300)
    plt.show()

    print(f"练习 4 柱状图已生成：{DISTRICT_COUNT_IMAGE}")

    print("\n练习 4：不同市区房源数量")
    for item in result:
        print(item["市区"], item["房源数量"])


# =========================
# 9. 程序入口：四个练习合成一个
# =========================

if __name__ == "__main__":

    print("当前爬取城市：", CITY)
    print("即将打开的网址示例：", f"https://{CITY}.anjuke.com/sale/p1-y1/?from=fangjia")

    # 练习 1：爬取厦门 500 条
    crawl_city(CITY, LIMIT)

    # 练习 1 表格：原始房源数据表
    export_house_list_csv(CITY)

    # 练习 2 表格：不同市区平均总价
    export_district_avg_price_table(CITY)

    # 练习 3 表格：不同户型数量
    export_house_type_count_table(CITY)

    # 练习 4 表格 + 图：市区房源数量
    export_district_count_table_and_chart(CITY)

    print("\n全部练习完成！")
    print("生成的文件如下：")
    print(HOUSE_LIST_CSV)
    print(DISTRICT_AVG_PRICE_CSV)
    print(HOUSE_TYPE_COUNT_CSV)
    print(DISTRICT_COUNT_CSV)
    print(DISTRICT_COUNT_IMAGE)
```
